Camada Gold - Análise de acidentes Rodoviários Pyspark/ DataBricks

```python
from pyspark.sql.functions import * 

#============================================
#Objetivo das Análises
#============================================

#1. Qual o poluente que mais influencia?
#2. Qual a Estacao que mais polue?
#3. Qual o Mês de maior poluição?
```

```python
#============================================
# Carregar tabela silver
#============================================

df_gold = spark.table("projeto_meio_ambiente.datambiente_2015_2021.silver_datambiente_15_21")
```

```python
#============================================
#Visualizar dados
#============================================

df_gold.show(5)
display(df_gold)
```

```python
#========================================
#Pergunta 1
#1. Qual o poluente que mais influencia
#========================================

Gold_Maior_Poluente = (
    df_gold.groupBy("Poluente")
    .agg(count("*").alias("Maior_Poluente"))
    .orderBy(col("Maior_Poluente").desc())                 
)

print("Maior Poluente")
display(Gold_Mais_Poluente)
```

```python
#=================================================
# Salvar tabela gold
#=================================================

Gold_Maior_Poluente.write\
    .format("delta")\
    .mode("overwrite")\
    .saveAsTable("projeto_meio_ambiente.datambiente_2015_2021.gold_datambiente_15_21")
```

```python
#========================================
#2. Qual a Estacao que mais polue?
#========================================

Gold_Mais_Poluente = (
    df_gold.groupBy("Estacao")
    .agg(count("*").alias("Mais_Polue"))
    .orderBy(col("Mais_Polue").desc())                 
)

print("Mais Polue")
display(Gold_Mais_Poluente)
```

```python
#=========================================
#salvar tabela Gold_Mais_Poluente
#========================================

Gold_Mais_Poluente.write\
    .format("delta")\
    .mode("overwrite")\
    .saveAsTable("projeto_meio_ambiente.datambiente_2015_2021.gold_estacao_polui")
```
